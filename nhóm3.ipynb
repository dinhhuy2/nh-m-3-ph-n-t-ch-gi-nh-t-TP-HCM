```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file CSV
df = pd.read_csv("cleanMey 1.csv")
print("=== 5 dòng đầu tiên của dữ liệu ===")
print(df.head())

# 2. Phân tích bằng List
gia_list = df["Price_Billion"].tolist()
print("\n=== Phân tích bằng List ===")
print("Giá trung bình:", sum(gia_list)/len(gia_list))
print("Giá cao nhất:", max(gia_list))
print("Giá thấp nhất:", min(gia_list))

# 3. Loại bỏ trùng lặp bằng Set (distinct)
quan_list = df["District"].tolist()
quan_distinct = set(quan_list)
print("\n=== Các quận có trong dữ liệu ===")
print(quan_distinct)

# 4. Phân tích bằng NumPy
gia_array = np.array(gia_list)
print("\n=== Phân tích bằng NumPy ===")
print("Trung bình:", np.mean(gia_array))
print("Độ lệch chuẩn:", np.std(gia_array))
print("Phân vị 25%:", np.percentile(gia_array, 25))
print("Phân vị 75%:", np.percentile(gia_array, 75))

# 5. Phân tích bằng Pandas
print("\n=== Thống kê cơ bản bằng Pandas ===")
print(df.describe())
print("\n=== Trung bình giá theo quận ===")
print(df.groupby("District")["Price_Billion"].mean())
print("\n=== Sắp xếp theo giá (giảm dần) ===")
print(df.sort_values(by="Price_Billion", ascending=False))

# 6. Biểu đồ trực quan
plt.figure(figsize=(10,6))
sns.barplot(x="District", y="Price_Billion", data=df, estimator=np.mean, errorbar=None, palette="viridis")
plt.title("Giá trung bình nhà đất theo quận (TP.HCM)")
plt.ylabel("Giá trung bình (Tỷ VND)")
plt.xlabel("Quận/Huyện")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
# 7 Boxplot - phân bố giá theo quận
plt.figure(figsize=(10,6))
sns.boxplot(
    x="District",
    y="Price_Billion",
    data=df
)

plt.title("Phân bố giá nhà đất theo quận (TP.HCM)")
plt.ylabel("Giá (Tỷ VND)")
plt.xlabel("Quận/Huyện")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
